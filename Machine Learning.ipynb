## Formula Linear Regression : 𝑦 = 𝛼 +β𝑥

- 𝑦 : response variable
- 𝑥 : explanatory variable / features
- 𝑦 : intercept / merupakan titik pada sumbu y, dimana garis linear yang terbentuk menabrak garis sumbu y atau dengan kata lain intecept adalah nilai sumbu y ketika sumbu x nya bernilai 0
- β : slope / nilai slope akan berpengaruh pada tingkat kemiringan dari garis linear yang terbentuk, dimana nilai slope 0 akan memberikan garis horizontal





```python
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
```

## Proses kalkukalasi dibalik nilai slope ( mencari nilai slope )

- Nilai slope pada linear regression bisa diperoleh dengan memanfaatkan formula berikut:

β = cov(𝑥,𝑦) / var(𝑥)    

- x sekumpulan features
- y sekumpulan nilai target



```python
print(f'X: \n{X}\n')
print(f'X flatten : {X.flatten()}\n')
print(f'y: {y}')
```

## Variance

```python
variance_x = np.var(X.flatten(), ddof=1) #ddof = degree of freedom

print(f'variance_x : {variance_x}')
```

## Covariance

```python
np.cov(X.flatten(),y)

# nilai covariance yang digunakan adalah yang dalam diagonal 22,65
```

```python
covariance_xy = np.cov(X.flatten(), y)[0][1]
ẍ
print(f'covariance: {covariance_xy}')
```

## Slope

```python
slope = covariance_xy / variance_x

print(f'slope : {slope}')
```

## Mencari nilai intercept

- Nilai intercept pada linear regression bisa diperoleh dengan memanfaatkan forumula berikut:

𝛼 = ȳ - βẍ

```python
intercept = np.mean(y) - slope * np.mean(X)

print(f'intercept: {intercept}')
```

## Teknik untuk mengevaluasi performa linear regression model

## Training & Testing dataset

```python
X_train = np.array([6,8,10,14,18]).reshape(-1,1)
y_train = np.array([7,9,13,17.5,18])

X_test = np.array([8,9,11,16,12]).reshape(-1,1)
y_test = np.array([11,8.5,15,18,11])
```

## Training simple linear regression model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

## Evaluasi Linear Regression Model dengan Coefficient of Determination atau R-Squared(R²)

- https://en.wikipedia.org/wiki/Coefficient_of_determination

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)

r_squared = r2_score(y_test, y_pred)

print(f'r_squared : {r_squared}')

# - bila hasil mendekati 1 semakin baik
# - semakin menjauhi 1 atau mendekati 0 maka semakin buruk
# - bila hasilnya negatif maka disebut buruk sekali
```

## Mencari nilai R-squared(R²)




- SSres = some square residual (proses perhitungan selisih antara setiap nilai y pada testing set dengan nilai prediksinya, lalu nilai selisihnya dipangkatkan 2 dan dijumalhkan secara keseluruhan. atau dengan kata lain kita menghitung nilai erornya atau akumulasi nilai erornya
- SStot = some square total (proses perhitungan selisih nilai untuk setiap nilai y pada testing set dengan nilai rata ratanya(ȳ) lalu dipangkatkan 2 dan diakumulasikan nilai yang ada.


## SSres

```python
ss_res = sum([(y_i - model.predict(x_i.reshape(-1,1))[0])**2
              for x_i, y_i in zip(X_test, y_test)])

print(f'ss_res : {ss_res}')
```

## SStot

```python
mean_y = np.mean(y_test)
ss_tot = sum([(y_i - mean_y)**2 for y_i in y_test])

print(f'SStot : {ss_tot}')
```

## R²

```python
r_squared = 1 - (ss_res / ss_tot)

print(f'R2 : {r_squared}')
```

# Classification dengan KNN ( K Nearest Neirest Neighbours )

- KNN adalah model machine learning yang dapat digunakan untuk melakukan prediksi berdasarkan kedekatan karakteristik dengan sejumlah tetangga terdekat.
- Prediksi yang dilakukan dapat diterapkan baik pada classification maupun regression tasks

## 1 Sample Dataset

```python
import pandas as pd

sensus = {
          'tinggi': [158, 170, 183, 191, 155, 163, 180, 158, 178],
          'berat': [64, 86, 84, 80, 49, 59, 67, 54, 67],
          'jk': ['pria','pria','pria','pria','wanita','wanita','wanita','wanita','wanita']}

sensus_df = pd.DataFrame(sensus)
sensus_df
```

## 2 Visualisasi Data

```python
import matplotlib.pyplot as plt

fig, ax =plt.subplots()
for jk, d in sensus_df.groupby('jk'):
    ax.scatter(d['tinggi'], d['berat'], label=jk)
    
plt.legend(loc='upper left')
plt.title('Sebaran data tinggi badan, berat badan, dan jenis kelamin')
plt.xlabel('Tinggi badan (cm)')
plt.ylabel('Berat badan (kg)')
plt.grid(True)
plt.show()
```

## 3 Preprocessing Dataset

```python
import numpy as np

X_train = np.array(sensus_df[['tinggi','berat']])
y_train = np.array(sensus_df[['jk']])

print(f'X_train : \n{X_train}\n')
print(f'y_train : {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train : \n{y_train}')
```

## 9 Metric Accuracy



- Matrix evaluasi pertama yang akan kita pelajari adalah akurasi
- Akurasi adalah proporsi dari instance pada training setyang di klasifikasikan secara benar atau yang berhasil diprediksi dengan tepat

- tp, tn = true positif(sesuatu yang bernilai positif telah dengan tepatdiprediksi sebagai positif oleh model), true negatif(sesuatu yang bernilai negatif telah dengan tepatdiprediksi sebagai negatif oleh model)

- keduanya merepresentasikan hasil prediksi / klasifikasi yang benar 
- contoh pada mengacu kasus saat ini adalah :
- tp = berarti model sudah dengan tepat memprediksi data pria sebagai pria dan data wanita sebagai wanita
- tn = berarti model sudah dengan tepat memprediksi data wanita sebagai bukan pria dan data pria bukan sebagai wanita


- fp, fn = false positif, false negatif
- keduanya merepresentasikan prediksi atau klasifikasi yang salah
- fp = berarti sesuatu yang bernilai negatif telah keliru diprediksi sebagai positif oleh model
- fn = berarti sesuatu yang bernilai positif telah keliru diprediksi sebagai negatif oleh model

- atau kalau pada mengacu pada contoh kasus ini :
- fp = berarti model sudah dengan keliru memprediksi data wanita sebagai pria dan data pria sebagai wanita
- sedangakan
- fn = berarti model sudah dengan keliru memprediksi data pria sebagai bukan pria dan data warga sebagai bukan wanita

## Mean Squared Error (MSE) atau Mean Squared Deviation (MSD)

- Metric Evaluation ketiga adalah Mean Squared Error (MSE) atau Mean Squared Deviation (MSD)
- Merupakan rata rata dari r kuadrat untuk prediksi 

- kita juga akan menghitung error 
- artinya disini kita juga akan menghitung selisih antara yi dan yi (head)
- yi merupakan nilai target dari testing set
- yi (head) merupakan nilai estimasi yang dihasilkan oleh model kita
- selisihnya bisa positif, bisa negatif
- nah untukmenghindari muncul nilai negatif pada MSE, selisih nilai yi-yi(head)kita pangkatkan 2
- lalu hasil pemangkatannya ini diakumulasikan untuk kita bagi dengan n (jumlah data point)

## Menerapkan Features Scaling pada KNN

```python
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import euclidean

ss = StandardScaler()

# training set

X_train = np.array([[158, 0], [170, 0], [183, 0], [191, 0], [155, 1], [163, 1], [180, 1], [158, 1], [170, 1]])

y_train = np.array([64, 86, 84, 80, 49, 59, 67, 54, 67])


# Test set

X_test = np.array([[168, 0], [180, 0], [160, 1], [169, 1]])
y_test = np.array([69, 92, 52, 67])
```

## Features Scalling (Standard Scaler)

```python

X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)

print(f'X_train_scaled : \n{X_train_scaled}\n')
print(f'X_test_scaled : \n{X_test_scaled}\n')
```

## Training & Evaluasi Model

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

K = 3
model = KNeighborsRegressor(n_neighbors=K)

model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)

print(f'MAE : \n{MAE}\n')
print(f'MSE : \n{MSE}\n')

# kalo kita bandingkan dengan MAE & MSE sebelumnya, kita lebih baik / lebih kecil (menghasilkan model 
# dengan kualitas atau perfoma yang lebih baik) setelah menerapkan Stantadrd scaling 
```

### Training Dataset

```python
pizza = {'diameter':[6,8,10,14,18],
         'n_topping':(2, 1, 0, 2, 0),
         'harga':[7,9,13,17.5,18]}

train_pizza_df = pd.DataFrame(pizza)
train_pizza_df
```

### Testing dataset

```python
pizza = {'diameter':[8,9,11,16,12],
         'n_topping':(2, 0, 2, 2, 0),
         'harga':[11,8.5,15,18,11]}

test_pizza_df = pd.DataFrame(pizza)
test_pizza_df
```

- perbedaan simple linear regression dengan multiple linear regression adalah 
- untuk simple kita hanya menggunakan 1 features aja untuk prediksi nilai
- pada multiple kita menggunakan lebih dari 1 feature untuk melakukan prediksi
- Formula multiple terlampir di gambar

```python
X_train = np.array(train_pizza_df['diameter']).reshape(-1,1)
y_train = np.array(train_pizza_df['harga'])

print(f'X_train : {X_train}')
print(f'y_train : {y_train}')
```
